# 5. Model Building

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB, MultinomialNB

from sklearn.model_selection import GridSearchCV
```

```python
model_params = {
    'svc': {
        'model': SVC(), #gamma='auto'
        'params': {
            'gamma': ['scale', 'auto'],
            'C':[1,10,20],
            'kernel': ['rbf','sigmoid']  # 'linear','poly',
            }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators':[50,100,150,200], 
            'criterion':['gini','entropy']
            }
    },
    'logistic_regression': {
        'model': LogisticRegression(),
        'params': {
            'solver':['lbfgs', 'liblinear', 'sag', 'saga', 'newton-cg',], 
            'C': [1,5,10],
            'penalty': ['l1', 'l2', 'elasticnet'], # 'none'
            'dual': [True, False],
            'fit_intercept': [True, False],
            'multi_class' : ['ovr', 'multinomial', 'auto']
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion':['gini','entropy'],
            'splitter': ['best','random'],
            'max_features': ['auto','sqrt','log2',None] 
        }
    },
    'KNeighbors': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors' : [2,3,5],
            'weights': ['uniform', 'distance'],
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
            'leaf_size': [10,20,30,40,50,60],
            'p' : [1,2]
        }
    },
    'BernoulliNB': {
        'model': BernoulliNB(),
        'params': {
            'alpha': [0,1],
            'fit_prior': [True,False]
        }
    },
}
```

```python
X = df.drop('Loan_Status', axis='columns')
y = df.Loan_Status.values
```

```python
def optimum_gridsearch(model_parameters, X, y):
    '''
    model_parameters: 'dict'
    X: features matrix
    y: target array
    '''
    scores = []
        
    for estimator,item in model_parameters.items():
        if item['params'] is None:  # to account for models with no tunable parameters.
            continue
        clf = GridSearchCV(item['model'], item['params'],
                           cv=15, return_train_score=False)
        clf.fit(X, y)
        scores.append({
            'model': estimator,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
```

```python
optimum = optimum_gridsearch(model_params, X, y)
```
